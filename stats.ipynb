```python
OWL2Bench_dbs = [
    {
        "path": "./datasets/OWL2Bench/OWL2Bench1/",
        "train_file": "_train_OWL2Bench1.owl",
        "test_file": "_test_OWL2Bench1.owl",
        "file": "OWL2Bench1",
    },
    {
        "path": "./datasets/OWL2Bench/OWL2Bench2/",
        "train_file": "_train_OWL2Bench2.owl",
        "test_file": "_test_OWL2Bench2.owl",
        "file": "OWL2Bench2",
    },
]
```

```python
for db_ in OWL2Bench_dbs:
    path = db_["path"]
    train_file = db_["train_file"]
    test_file = db_["test_file"]

    df_train = load_ore_files(path + train_file)
    df_test = load_ore_files(path + test_file)

    # type rule - train
    o = df_train[df_train["p"] == "ClassAssertion"]["o"]
    print(len(df_train[(df_train["p"] == "SubClassOf") & (df_train["s"].isin(o))]))

    # type rule - test
    o = df_test[df_test["p"] == "ClassAssertion"]["o"]
    print(len(df_test[(df_test["p"] == "SubClassOf") & (df_test["s"].isin(o))]))

    # subclass rule - train
    o = df_train[df_train["p"] == "SubClassOf"]["o"]
    print(len(df_train[(df_train["p"] == "SubClassOf") & (df_train["s"].isin(o))]))

    # subclass rule - test
    o = df_test[df_test["p"] == "SubClassOf"]["o"]
    print(len(df_test[(df_test["p"] == "SubClassOf") & (df_test["s"].isin(o))]))
```

```python
for db_ in OWL2Bench_dbs:
    path = db_["path"]
    train_file = db_["train_file"]
    test_file = db_["test_file"]
    file = db_["file"]

    (
        g_train,
        g_train_filter_subclass,
        g_train_filter_assertion,
        g_test,
        g_test_filter_subclass,
        g_test_filter_assertion,
    ) = load_ore_graphs(path, train_file, test_file)

    print(f"# Train - Subclass Edges: {g_train_filter_subclass.number_of_edges()}")
    print(f"# Test - Subclass Edges: {g_test_filter_subclass.number_of_edges()}")

    print(f"# Train - Assertion Edges: {g_train_filter_assertion.number_of_edges()}")
    print(f"# Test - Assertion Edges: {g_test_filter_assertion.number_of_edges()}")
    print()
```

```python
for db_ in CLG_dbs:
    path = db_["path"]
    train_file = db_["train_file"]
    test_file = db_["test_file"]

    df_train = load_clg_files(path + train_file)
    df_test = load_clg_files(path + test_file)

    # type rule - train
    o = df_train[df_train["p"] == "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>"][
        "o"
    ]
    print(
        len(
            df_train[
                (df_train["p"] == "<http://www.w3.org/2000/01/rdf-schema#subClassOf>")
                & (df_train["s"].isin(o))
            ]
        )
    )

    # type rule - test
    o = df_test[df_test["p"] == "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>"][
        "o"
    ]
    print(
        len(
            df_test[
                (df_test["p"] == "<http://www.w3.org/2000/01/rdf-schema#subClassOf>")
                & (df_test["s"].isin(o))
            ]
        )
    )

    # subclass rule - train
    o = df_train[df_train["p"] == "<http://www.w3.org/2000/01/rdf-schema#subClassOf>"][
        "o"
    ]
    print(
        len(
            df_train[
                (df_train["p"] == "<http://www.w3.org/2000/01/rdf-schema#subClassOf>")
                & (df_train["s"].isin(o))
            ]
        )
    )

    # subclass rule - test
    o = df_test[df_test["p"] == "<http://www.w3.org/2000/01/rdf-schema#subClassOf>"][
        "o"
    ]
    print(
        len(
            df_test[
                (df_test["p"] == "<http://www.w3.org/2000/01/rdf-schema#subClassOf>")
                & (df_test["s"].isin(o))
            ]
        )
    )
```
